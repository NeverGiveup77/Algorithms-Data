```python
import random

A = [random.randint(0, 20) for i in range(20)]
A.sort()

def binary_search(arr, key):
    left = 0
    right = len(arr) - 1
    while left <= right:
        mid = (right + left) // 2
        if key > arr[mid]:
            left = mid + 1
        elif key < arr[mid]:
            right = mid - 1
        else:
            return mid
    return "No results"


print(binary_search(A, 10))
```

```python
def fibonaciian_search(arr, x, n):

    # Initialize fibonacci numbers
    fib_mm2 = 0
    fib_mm1 = 1
    fib_mm = fib_mm2 + fib_mm1

    # fib_mm is going to store the smallest
    # Fibonacci number greater that or equal to n
    while fib_mm < n:
        fib_mm2 = fib_mm1
        fib_mm1 = fib_mm
        fib_mm = fib_mm2 + fib_mm1

    # Marks the eliminated range from front
    offset = -1

    # while there are elements to be inspected
    # note that we compare arr[fib_mm2] with x
    # when fib_mm becomes 1, fib_mm2 becomes 0
    while fib_mm > 1:

        # Check if fib_mm2 is a valid location
        i = min(offset + fib_mm2, n - 1)

        # if x is greater than the value at
        # index fib_mm2, cut the subarray array
        # from offset to i
        if arr[i] < x:
            fib_mm = fib_mm + fib_mm1
            fib_mm1 = fib_mm1 + fib_mm2
            fib_mm2 = fib_mm1 - fib_mm2
            offset = i

        # If x is less than the value at
        # index fib_mm2, cut the subarray
        # after i + 1
        elif arr[i] > x:
            fib_mm = fib_mm1
            fib_mm1 = fib_mm2
            fib_mm2 = fib_mm - fib_mm1

        else:
            return i
    # comparing the last element with x */
    if fib_mm and arr[n - 1] == x:
        return n - 1

    # element not found
    return -1

# Driver Code
import random
arr = [random.randint(1,25) for _ in range(20)]
arr.sort()
n = len(arr)
x = 14

print(arr)
ind = fibonaciian_search(arr, x, n)
if ind>=0:
  print("Found at index:",ind)
else:
  print(x,"isn't present in the array")
```

```python
from random import randint
import time

start = time.time()
class Tree:
    def __init__(self, val):
        self.left = None
        self.right = None
        self.val = val

    def insert(self, val):
        if self.val:
            if val < self.val:
                if self.left is None:
                    self.left = Tree(val)
                else:
                    self.left.insert(val)
            else:
                if self.right is None:
                    self.right = Tree(val)
                else:
                    self.right.insert(val)
        else:
            self.val = val

    def search(self, val):
        if self.val == val:
            print("Success!", val, "найдено!")
            return self
        else:
            if val < self.val:
                if self.left is not None:
                    return self.left.search(val)
                else:
                    print(val, "not found!")
            elif val > self.val:
                if self.right is not None:
                    return self.right.search(val)
                else:
                    print(val, "not found!")

    def minValueNode(self):
        current = self

        # loop down to find the leftmost leaf
        while current.left is not None :
            current = current.left

        return current

    def deleteNode(self, val):
        if self is None:
            return self
        if val < self.val:
            self.left = self.left.deleteNode(val)
        elif val > self.val:
            self.right = self.right.deleteNode(val)
        else:
            if self.left is None:
                temp = self.right
                self = None
                return temp
            elif self.right is None:
                temp = self.left
                self = None
                return temp
            temp = self.right.minValueNode()
            self.val = temp.val
            self.right = self.right.deleteNode(temp.val)
        return self

    def collectValues(self, obj):
        arr = []
        if obj is not None:
            arr = self.collectValues(obj.left)
            obj.left = None
            arr.append(obj.val)
            arr += self.collectValues(obj.right)
            obj.right = None
        return arr

    def printTree(self):
        if self is not None:
            print("Root: ", self.val)
            if self.left is not None:
                # print("Root: ", self.val)
                print("L:", self.left.val)
                self.left.printTree()
            if self.right is not None:
                # print("Root: ", self.val)
                print("R:", self.right.val)
                self.right.printTree()

    def inorder(self):
        if self.val:
            if self.left is not None:
                self.left.inorder()
            print(self.val)
            if self.right is not None:
                self.right.inorder()


arr = [randint(2, 6) for _ in range(10)]
print(arr)
# arr = [8, 0, 2, 9, 9, 6]
root = Tree(arr[0])
for x in arr[1::]:
    root.insert(x)

root.inorder()
print("after deletion")
root.deleteNode(5)
root.inorder()

print(time.time() - start)
```

```python
import random
A = [random.randint(1,30) for _ in range(20)]
A.sort()

def interpolation_search(arr, x):
    lo = 0
    hi = len(arr) - 1
    # The idea of formula is to return higher value of pos
    # when element to be searched is closer to arr[hi]. And
    # smaller value when closer to arr[lo]
    while True:
        arr_len = len(arr)
        pos = lo + (x-arr[lo])*(hi-lo) // (arr[hi]-arr[lo])
        if arr[pos] == x:
            return pos
        elif arr[pos] > x:
            arr = arr[:pos]
            hi = len(arr) - 1
        elif arr[pos] < x:
            arr = arr[pos+1:]
            hi = len(arr) - 1
        if len(arr) == 0:
            return -1

res = interpolation_search(A, 15)
print(res)
```

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.T = [None] * size

    def h(self, val, i = 0):
        if isinstance(val, str):
            num = 0
            for j in val:
                num += ord(j)
            return (num + i) % self.size
        return (val + i) % self.size

    def hash_insert(self, k):
        for i in range(len(self.T)):
            j = self.h(k, i)
            if self.T[j] is None or self.T[j] == "deleted":
                self.T[j] = k
                return

    def hash_search(self, k):
        for i in range(len(self.T)):
            j = self.h(k, i)
            if self.T[j] is None:
                # print("no such element")
                return -1
            elif self.T[j] == k:
                # print("found element {0} in pos = {1}".format(k, j))
                return j
        # print("no such element")
        return -1

    def hash_delete(self, k):
        j = self.hash_search(k)
        if self.T[j] == k:
            self.T[j] = "deleted"
            print("element deleted successfully")

# Driver code
import random
import time
n = 0
start = time.time()
while n < 100:
    arr = [random.randint(-1000, 1000) for _ in range(1000)]
    ht = HashTable(len(arr))
    for x in arr:
        ht.hash_insert(x)

    n += 1

print(time.time() - start)
```

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.T = [None] * size
        self.rand_index = []

    def h(self, val):
        i = random.randrange(0, self.size)
        if isinstance(val, str):
            num = 0
            for j in val:
                num += ord(j)
            self.rand_index.append((num + i) % self.size)
            return (num + i) % self.size
        self.rand_index.append((val + i) % self.size)
        return (val + i) % self.size

    def hash_insert(self, k):
        j = self.h(k)
        if self.T[j] is None or self.T[j] == "deleted":
            self.T[j] = k
            return

    def hash_search(self, k):
        for i in self.rand_index:
            if self.T[i] == k:
                # print("found element {0} in pos = {1}".format(k, i))
                return i
        # print("no such element")
        return -1

    def hash_delete(self, k):
        j = self.hash_search(k)
        if self.T[j] == k:
            self.T[j] = "deleted"
            print("element deleted successfully")

# Driver code
import random
import time
n = 0
start = time.time()
while n < 100:
    arr = [random.randint(-1000, 1000) for _ in range(1000)]
    ht = HashTable(len(arr))
    for x in arr:
        ht.hash_insert(x)

    n += 1

print(time.time() - start)
```

```python
from random import randint
import builtins
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None


class LinkedList:
    def __init__(self):
        self.headval = None

    def insert(self, v):
        if self.headval is None:
            self.headval = Node(v)
            return
        l_val = self.headval
        while l_val.nextval:
            l_val = l_val.nextval
        l_val.nextval = Node(v)

    def remove(self, x):
        headval = self.headval
        if headval is not None:
            if headval.dataval == x:
                self.headval = headval.nextval
                return 1

        while headval.dataval != x and headval is not None:
            prev = headval
            headval = headval.nextval
            if headval is None:
                return -1

        prev.nextval = headval.nextval
        del headval
        return 1

    def search(self, x):
        v = self.headval
        j = 0
        while v is not None:
            if v.dataval == x:
                return j
            v = v.nextval
            j += 1
        return -1

    def print_list(self):
        printval = self.headval
        list = ""
        while printval is not None:
            list += str(printval.dataval) + " -> "
            printval = printval.nextval
        return list.rstrip(" -> ")

class HashTable:
    def __init__(self, size):
        self.size = size
        self.T = [ LinkedList() for _ in range(size) ]

    def h(self, k):
        return k % len(self.T)

    def chained_hash_insert(self, x):
        self.T[self.h(x)].insert(x)

    def chained_hash_search(self, x):
        print("{} if found in T[{}] on pos {}".format(x, self.h(x), self.T[self.h(x)].search(x)))

    def chained_hash_delete(self, x):
        print("{} element is deleted = {}".format(x, self.T[self.h(x)].remove(x)))

    def print_table(self):
        [print("T[{}] = {}".format(i, self.T[i].print_list())) for i in range(self.size)]

# Driver code
import random
import time
n = 0
start = time.time()
arr = [random.randint(0,20) for _ in range(10)]
ht = HashTable(len(arr)//3)
for i in arr:
    ht.chained_hash_insert(i)

# ht.print_table()
# ht.chained_hash_search(10)
# ht.chained_hash_delete(10)
# ht.print_table()
# ht.chained_hash_search(10)
while n < 100:
    arr = [random.randint(-1000,1000) for _ in range(1000)]
    ht = HashTable(len(arr)//3)
    for i in arr:
        ht.chained_hash_insert(i)

    n += 1

print(time.time() - start)
```

```python
import time

start = time.time()
board = [[0 for j in range(8)] for i in range(8)]
def setQueen(i, j):
    for x in range(8):
        board[x][j] += 1
        board[i][x] += 1
        foo = j - i + x
        if 0 <= foo < 8:
            board[x][foo] += 1
        foo = j + i - x
        if 0 <= foo < 8:
            board[x][foo] += 1
    board[i][j] = -1


def resetQueen(i, j):
    for x in range(8):
        board[x][j] -= 1
        board[i][x] -= 1
        foo = j - i + x
        if 0 <= foo < 8:
            board[x][foo] -= 1
        foo = j + i - x
        if 0 <= foo < 8:
            board[x][foo] -= 1
    board[i][j] = 0


def tryQueen(i):
    result = False
    for j in range(8):
        if board[i][j] == 0:
            setQueen(i, j)
            if i == 7:
                result = True
            else:
                result = tryQueen(i + 1)
                if not result:
                    resetQueen(i, j)
        if result:
            break
    return result


tryQueen(0)
for i in range(8):
    for j in range(8):
        if board[i][j] == -1:
            print("1", end="")
        else:
            print("0", end = "")
    print()

finish = time.time()
print()
print(finish - start)
```
